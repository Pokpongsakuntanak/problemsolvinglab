```python
import time
def sum_even_odd(n):
    lim = n

    num_even = lim // 2
    sum_even = num_even * (num_even + 1)

    num_odd = (lim + 1) // 2
    sum_odd = num_odd ** 2

    return sum_odd , sum_even

try:
    n = int(input("Enter the number do you want:"))
except ValueError:
    print("ใส่ให้ถูกพี่ขก.แก้โค้ด")

start = time.time()
answer_v1 = sum_even_odd(n)
end = time.time()
time_v1 = (end - start) * 1000 # แปลงเป็นมิลลิวินาที
print(f'answer: {answer_v1}')
print(f'time: {time_v1:.4f} ms')
```
